```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import glob
```

```python
data = pd.read_csv('/content/single_manager_fund.csv',encoding='latin1')
data = data[data['n_fund_per_manager']>= 2] #Filter out managers who only have one fund and can't caculate the correlation
data = data[~data['mgr_name'].isin(['Not Managed'])]
data = data.loc[: ,[ 'crsp_portno','m_date', 'fund_name','weight_mret','mgr_name']]  # Delete irrelevant columns
```

```python
single = pd.DataFrame()
for i,group in  data.groupby("mgr_name"):
  single = group
  single.to_csv(path_or_buf=i+".csv")  # Split the dataset based on the manager's name， then save the result in csv.file
  print('*'*13+i+'*'*13+'\n', group)
```

```python
corr_result = []
fund_name_1 = []
fund_name_2 = []
manager_name = []

path = "/content/*.csv"
csv_files = glob.glob(path)
for file in csv_files:
  data_corr = pd.read_csv(file)  # use loop to read all csv file
  print(file)
  corrs = data_corr.pivot(index = "m_date", columns = "crsp_portno", values = "weight_mret") # convert to pivot table
  row, column = corrs.shape
  fund_range = list(corrs.columns)
  corr = corrs.corr().values
  for i in range(column):
    for j in range(i+1, column):
        if row >= 12:
          corr_result.append(corr[i,j]), 
          fund_name_1.append(fund_range[i]),
          fund_name_2.append(fund_range[j]),
          manager_name.append(file)
          print( fund_range[i], ' ', fund_range[j], ' ', corr[i,j], file)
  print("**********")
```

```python
final_result = pd.DataFrame(
    {
        'fund name 1:' : fund_name_1,
        'fund name 2:' : fund_name_2,
        'correlation:' : corr_result,
        'manager_name:' : manager_name
    }
)  # Merge four list in one dataframe
```
