```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import glob
```

```python
data = pd.read_csv('/content/single_manager_fund.csv',encoding='latin1')
```

```python
data = data[data['n_fund_per_manager']>= 2] #Filter out managers who only have one fund and can't caculate the correlation
data = data[~data['mgr_name'].isin(['Not Managed'])]
```

```python
data = data.loc[: ,[ 'crsp_portno','m_date', 'fund_name','weight_mret','mgr_name']]  # Delete irrelevant columns
```

```python
data
```

```python
single = pd.DataFrame()
seperate= pd.DataFrame()
for (i, j),group in  data.groupby(["mgr_name","fund_name"]):
  single = group
  single.to_csv(path_or_buf=i+".csv")
  print('*'*13+i+j+'*'*13+'\n', group)
```

```python
single = pd.DataFrame()
for i,group in  data.groupby("mgr_name"):
  single = group
  single.to_csv(path_or_buf=i+".csv")
  print('*'*13+i+'*'*13+'\n', group)
```

```python
path = "/content/*.csv"
csv_files = glob.glob(path)
for file in csv_files:
  data_corr = pd.read_csv(file)
  print(file)
  corrs = data_corr.pivot(index = "m_date", columns = "crsp_portno", values = "weight_mret")
  corrs = corrs.dropna(axis = 0, how = 'any')
  corrs = corrs.corr(method='pearson')
  print(corrs)
  print("**********")
```
