```python
import pandas as pd
import numpy as np
import statsmodels.api as sm # simple linear regression model as there is only one independent variable
```

```python
data=pd.read_csv('/content/excess_fund_return.csv')
data
```

```python
feature_cols = ['mktrf']
```

```python
X = data[feature_cols]# Feature
y = data.efr #target variable
```

```python
reg = sm.OLS(y, X).fit()
```

```python
reg_result = reg.resid
```

```python
reg_result
```

```python
reg_result.to_csv(path_or_buf="d.csv")
```
