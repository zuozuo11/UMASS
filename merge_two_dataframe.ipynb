```python
data_1 = pd.read_csv('/content/single_manager_fund.csv',encoding='latin1')
data_1 = data_1[data_1['n_fund_per_manager']>= 2] #Filter out managers who only have one fund and can't caculate the correlation
data_1 = data_1[~data_1['mgr_name'].isin(['Not Managed'])]
data_1 = data_1.loc[: ,[ 'crsp_portno','m_date', 'fund_name','weight_mret','mgr_name']]  # Delete irrelevant columns
```

```python
merge_result = pd.merge(data_1,data_2, left_on = 'm_date', right_on = 'adjust_date' )
```
