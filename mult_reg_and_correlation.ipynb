```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_csv('/content/for_mult_reg.csv',encoding='latin1')
```

```python
data
```

```python
feature_cols = ['mktrf','hml','smb','umd']
```

```python
X = data[feature_cols]# Feature
y = data.efr #target variable
```

```python
reg = sm.OLS(y, X).fit()
reg.summary()
```

```python
data['res']=pd.DataFrame(reg.resid)
```
